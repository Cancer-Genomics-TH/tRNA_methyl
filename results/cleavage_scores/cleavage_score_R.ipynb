```r
in_fn <- "cleavage_scores_bo_seq.tsv"
cols_to_read = c("isotype", "start", "cleavage_score")

data <- read_delim(in_fn, col_select = cols_to_read)

data
```

```r
trnas <- unique(data$isotype)
trnas <- sort(trnas)
trnas
```

```r
for (i in 1:length(trnas)) {
  # isotype
  iso <- trnas[i]
  print( iso)
  iso_data <- data[data$isotype==iso, ]
  print(iso_data)
  # calculate cleavage score
  starts <- sort(unique(iso_data$start))
  starts <- starts[starts > 9]
  starts <- starts[starts < 61]
  print(starts)
  pdf.file <- paste("cleavage_score_2023_07_27/", iso, ".cleavege_corrected_2023_07_27a.pdf", sep="")
  pdf(pdf.file, width=4, height=4)
  plot(iso_data$start, iso_data$cleavage_score, type="h", 
       lwd=3, col=colors[10],
       main=iso, cex.main=1.6, 
       ylab="Cleavage score", 
       xlab="tF start site",  
       ylim=c(-10, 10),  
       xlim=c(10, 60), 
       cex.lab=1.2, 
       las=2, 
       cex.axis=1.1)
  dev.off()
  
}
 
```

```r
 cleavage_score = rep(0, length(starts))
  for(j in 1:length(starts)) {
      xs = iso_data[iso_data$start_in_trna == starts[j], ]
      zs = colSums(xs[,3:6])
      
      if ((z[2]+z[4]) > 0 & (zs[2]+zs[4]) > 0 & 
          (z[1]+z[3]) > 9 & (zs[1]+zs[3]) > 0) {
        cleavage_score[j] = log2(
            ((zs[2]+zs[4])/(z[2]+z[4]))/
            ((zs[1]+zs[3])/(z[1]+z[3]))
        )
      }
  }
  
  # plot cleavage score for isotype
  pdf.file <- paste("cleavage_score/", iso, ".cleavage_score.2023_02_22a.pdf", sep="")
  pdf(pdf.file, width=4, height=4)
  plot(starts, cleavage_score, type="h", 
       lwd=3, col=colors[10],
       main=iso, cex.main=1.6, 
       ylab="Cleavage score", 
       xlab="tF start site",  
       ylim=c(-10, 10),  
       xlim=c(10, 60), 
       cex.lab=1.2, 
       las=2, 
       cex.axis=1.1)
  dev.off()
  
  
}
  
```
